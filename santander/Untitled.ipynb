```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
train = pd.read_csv('data/train.csv', index_col=0)
train.head()
```

```python
train.describe()
```

### There are missing values?

```python
train.isnull().values.any()
```

### Any column is all zeros?

```python
total_zero_cols = 0
for col, vals in train.iteritems():
    if vals.sum() == 0:
        total_zero_cols += 1

print(total_zero_cols)
```

### Remove all zeros features

```python
train = train.loc[:, (train != 0).any(axis=0)]
```

### Rename features

```python
train.columns = ['target'] + list(range(1, 4736))
train.head()
```

### Features Correlation

```python
corr = train.corr()
```

```python
corr_matrix = corr.abs()
high_corr_var=np.where(corr_matrix>0.7)
high_corr_var=[(corr_matrix.columns[x],corr_matrix.columns[y]) for x,y in zip(*high_corr_var) if x!=y and x<y]
```

```python
to_delete = list(dict(high_corr_var).values())
train = train.drop(to_delete, axis=1)
```

### SelectKBest Features

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression

X = train.drop('target', axis=1)
y = train.target

X = SelectKBest(mutual_info_regression, k=2000).fit_transform(X, y)
X.head()
```

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=500)
X_pca = pca.fit_transform(X)
X_pca.shape
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=42
)
```

### Checkpoint

```python
import pickle

data = {
    "X_train": X_train,
    "X_test": X_test,
    "y_train": y_train,
    "y_test": y_test,
    "X": X
}

with open('data/prepared_data.pkl', 'wb') as f:
    pickle.dump(data, f)
```

### Linear Regression Test

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_log_error

lr = LinearRegression()
lr.fit(X_train, y_train)
```
